```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import getpass
```

```python
client_id = str(getpass.getpass('client_id?'))
```

```python
client_secret = str(getpass.getpass('client_secret?'))
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="",
                                                           client_secret=""))
```

```python
playlist1 = sp.user_playlist_tracks("Neil T", "3DI5LujjXYYks2Z470bXt8")
```

```python
len(playlist1['items'])
```

```python
playlist1['items'][0]['track']['id']

for item in playlist1['items']:
    print(item['track']['id'])

my_mini_playlist = []
for item in playlist1['items']:
    my_mini_playlist.append(item['track']['id'])
```

```python
my_mini_playlist
```

```python
sp.audio_features(my_mini_playlist)
```

```python
audio_features_mini = pd.DataFrame(sp.audio_features(tracks=my_mini_playlist))
```
