```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import getpass
```

```python
client_id = str(getpass.getpass('client_id?'))
```

```python
client_secret = str(getpass.getpass('client_secret?'))
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="...",
                                                           client_secret="..."))
```

```python
playlist2 = sp.user_playlist_tracks("Spotify", "37i9dQZF1DX6KItbiYYmAv")
```

```python
len(playlist2['items'])
```

```python
playlist2['items'][0]['track']['id']

for item in playlist2['items']:
    print(item['track']['id'])

hot_acoustics = []
for item in playlist2['items']:
    hot_acoustics.append(item['track']['id'])
```

```python
sp.audio_features(hot_acoustics)
```

```python
audio_features_hot_acoustics = pd.DataFrame(sp.audio_features(tracks=hot_acoustics))
```
